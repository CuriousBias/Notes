```python
from dataclasses import dataclass
import matplotlib.pyplot as plt
```

```python
@dataclass
class optic:
    name: str
    weight: float
    mag: list
    fov: list
    obj_dia: float = None
```

```python
# convert linear fov -> angular fov: linear/1.75 = angular (as long as units of linear are the same)
# metic: linear/1.75, imperial/5.25

tars = optic(name='tars', weight = 47, mag = [3, 15], fov = [7.150, 1.430])
tenmile318 = optic(name='tenmile3-18', weight=25.7, mag=[3, 18], fov=[34.7/5.25, 5.8/5.25])

ultra_bright = optic(name='ultra bright', weight = 37, mag = [4.45, 16], fov = [9.4/1.75, 2.3/1.75])
exos = optic(name= 'exos', weight=32.4, mag = [3, 21], fov = [13.0/1.75, 1.9/1.75])

x5i318 = optic(name= 'x5i3-18', weight=30, mag = [3, 18], fov = [5.8, 1.2])
x5i525 = optic(name= 'x5i5-25', weight=32, mag = [5, 25], fov = [4, .8])
z8i56 = optic(name='Z8i56', weight=25.6, mag=[2.3, 18], fov=[10.6, 1.3])
z8i50 = optic(name='Z8i50', weight=23.8, mag=[2, 16], fov=[12.0, 1.5])
z6i = optic(name='z6i', weight=23.1, mag=[2.5, 15], fov=[9.4, 1.5])

v6 = optic(name= 'v6', weight=22.2, mag = [3, 18], fov = [7.1, 1.2])
v8 = optic(name= 'v8', weight=28, mag = [2.8, 20], fov = [8.8, 1.2])
lrp318 = optic(name= 'lrp3-18', weight=32, mag = [3, 18], fov = [5.7, 1.1])

infinity320 = optic(name='infinity3-20', weight=30.3, mag=[3.12,22.3], fov=[14.68/1.75, 1.9/1.75])

magnus216 = optic(name='magnus2-16', weight=30.9, mag=[2.67, 16], fov=[18.91/1.75, 2.6/1.75])

khales318i = optic(name='k3-18i', weight=33.2, mag=[3.5, 18], fov =[27.8/5.25, 5.5/5.25])

mark5318 = optic(name='mark5hd3-18', weight=26, mag=[3.6, 18], fov=[5.4, 1.1])
mark5225 = optic(name = 'mark5hd2-25', weight=30, mag=[5, 25], fov = [4, .8])

m7xi428 = optic(name='m7Xi4-28', weight=33.5, mag=[4,28], fov=[9/1.75, 1.42/1.75])
m7xi320 = optic(name='m7Xi3-20', weight=32.5, mag=[3,20], fov=[12/1.75, 1.78/1.75])

tangent315 = optic(name='tangent3-15', weight=27.7, mag=[3,15], fov =[7.3,1.6])
zc420 = optic(name='zerocomp420', weight=34.8, mag=[4, 20], fov=[9.3/1.75, 2/1.75])
```

```python
optics = exos, v8, z8i50, z8i56, magnus216

fig, ax = plt.subplots(1, 1, figsize= (15, 10))
fig.patch.set_facecolor('white')

for _ in optics:
    ax.plot(_.mag, _.fov, label = _.name)
ax.set_title('Optics FOV Comparison')
ax.set_xlabel('Magnification [x]')
ax.set_ylabel('Field of View [Degree]')
ax.legend()

plt.savefig('Optics_comp.png')
plt.show()
```
