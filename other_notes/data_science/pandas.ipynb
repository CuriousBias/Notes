#### Pivots
Reorganize data

```python
data = [[1001, 'cycle_count', 1], [1002, 'cycle_count', 2], [1003, 'cycle_count', 3],
        [1001, 'time', 1.1], [1002, 'time', 1.3], [1003, 'time', .9],
        [1001, 'force', 15], [1002, 'force', 16], [1003, 'force', 18],]

df_01 = pd.DataFrame(data, columns=['epoch_ms', 'sig_name', 'sig_value'])
df_01
```

```python
df_01_pivot = df_01.pivot_table(index=df_01['epoch_ms'], columns=["sig_name"], values=["sig_value"], aggfunc=np.max).fillna(method='ffill').dropna()
df_01_pivot
```

```python
# second style of pivot. More useful!
df_01_pivot_2 = df_01.pivot(index='epoch_ms', columns='sig_name', values='sig_value').reset_index()
df_01_pivot_2.fillna(method='bfill', inplace=True)
df_01_pivot_2
```
